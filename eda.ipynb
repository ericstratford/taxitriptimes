```python
coords_list = []
for coordinates in taxi["POLYLINE"]:
    coordinates = coordinates[1:-1]
    coordinates = coordinates.split(']')
    for coord in coordinates[:-1]:
        coord = coord.replace('[', '')[1:]
        coords = coord.split(',')
        coords_list.append((float(coords[0]), float(coords[1])))

longitude = [coord[1] for coord in coords_list]
latitude = [coord[0] for coord in coords_list]
```

### Adding Ride Duration
This code taken from discussion adds the ride duration to the table (the prediction we are looking for)

```python
# Over every single 
def polyline_to_trip_duration(polyline):
  return max(polyline.count("[") - 2, 0) * 15

# This code creates a new column, "LEN", in our dataframe. The value is
# the (polyline_length - 1) * 15, where polyline_length = count("[") - 1
ride_duration = taxi["POLYLINE"].apply(polyline_to_trip_duration)
```

```python
taxi_time = taxi.copy()
taxi_time['RIDE_DURATION'] = ride_duration
taxi_time
```

```python
taxi_time
```

# Distribution of Travel Time

```python
ride_mean = taxi_time['RIDE_DURATION'].mean()
ride_std = taxi_time['RIDE_DURATION'].std()
```

```python
def standarize(col):
    return (col - ride_mean)/ride_std
taxi_time['RIDE_DURATION_STANDARD'] = taxi_time['RIDE_DURATION'].apply(standarize)
```

```python
log_plot = taxi_time.plot(kind = 'hist', y = 'RIDE_DURATION',log=True, bins=20, title = 'Ride Distribution Logged')
log_plot.set_xlabel("Log Ride Duration")
log_plot.set_ylabel("y label")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader

# Define the RNN model
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(RNN, self).__init__()
        self.hidden_size = hidden_size
        self.rnn = nn.RNN(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out, _ = self.rnn(x)
        out = self.fc(out[:, -1, :])  # Take the last output of the sequence
        return out

# Define the hyperparameters
input_size = 1  # Size of the input feature vector (including datetime as floats)
hidden_size = 5  # Number of hidden units in the RNN layer
output_size =1  # Desired output size

# Create the RNN model
model = RNN(input_size, hidden_size, output_size)

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Prepare the data
X = X  # Input tensor with shape (num_samples, input_size)
y = y  # Target tensor with shape (num_samples, output_size)
batch_size = 32  # Number of samples in each batch
sequence_length = 1  # Number of time steps or length of the sequence

# Reshape the input tensor to (batch_size, sequence_length, input_size)
#X = X.view(-1, sequence_length, input_size)

# Convert the input tensor to the appropriate data type
X = X.float()  # Assuming the input tensor is of type float

# Convert the target tensor to the appropriate data type
y = y.float()  # Assuming the target tensor is of type float

# Create a dataset and data loader
dataset = torch.utils.data.TensorDataset(X, y.unsqueeze(1))
data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Train the model
num_epochs = 10
model.train()
for epoch in range(num_epochs):
    for inputs, labels in data_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

    print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, loss.item()))
```
